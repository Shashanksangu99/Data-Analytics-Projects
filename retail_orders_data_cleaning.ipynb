```python
#extractin the csv file from zipfile
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() # extract file to dir
zip_ref.close()
```

```python
#import pandas library
import pandas as pd

#reading the file as pandas dataframe
orders_df=pd.read_csv('/content/orders.csv')
display(orders_df)
```

```python
#formatting the columns by converting them to lower case and adding _ to them like converting Order ID to order_id

orders_df.columns

orders_df.columns=orders_df.columns.str.lower()

orders_df.columns=orders_df.columns.str.replace(" ","_")

orders_df.head(5)
```

```python
#creating new calculated columns discount,sale_price,profit

orders_df['discount']=orders_df['list_price']*orders_df['discount_percent']*0.1

orders_df['sale_price']=orders_df['list_price']-orders_df['discount']

orders_df['profit']= orders_df['sale_price']-orders_df['cost_price']


orders_df
```

```python
#checking all the datatypes and converting if there are any mistakes

orders_df.dtypes
```

```python
#converting order_date to date format
orders_df['order_date']=pd.to_datetime(orders_df['order_date'],format='%Y-%m-%d')

orders_df.dtypes
```

```python
#dropping the columns cost_price, list_price and discount_percent columns
orders_df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True,errors='ignore')

orders_df
```

```python
#Saving all the. changes into a new modified_orders.csv file to download it and open in Azure Data Studio to do some analysis
orders_df.to_csv('modified_orders.csv', index=False)
```
